## ü§† Let's get fancy

```python
# define yml object/class
import os
import textwrap
import openai
class YML_AI(object):
  #
  # initialize the object
  def __init__(self, name="Pluto", is_verbose=True,*args, **kwargs):
    super(YML_AI, self).__init__(*args, **kwargs)
    self.author = "Duc Haba"
    self.version = 1.0
    self.name = name
    if (is_verbose):
      self._ph()
      self._pp("Hello from class", f"{self.__class__} Class: {self.__class__.__name__}")
      self._pp("Code name", self.name)
      self._pp("Author is", self.author)
      self._ph()
    #
    return
  #
  # pretty print output name-value line
  def _pp(self, a, b):
    print("%28s : %s" % (str(a), str(b)))
    return
  #
  # pretty print the header or footer lines
  def _ph(self):
    print("-" * 28, ":", "-" * 28)
    return
  #
  # parse the answer
  def get_answer(self, resp, index=0):
    return resp.get('choices')[index].get('text')
  # print out the answer
  def print_answer(self, resp, index=0,is_print_json=True,is_return_val=False):
    print('----------')
    print('The Answer')
    print('----------')
    bdata = None
    rdata = self.get_answer(resp, index)
    print(textwrap.fill(rdata, width=72, replace_whitespace=False))
    if (is_print_json):
      print('----------')
      print('JSON Response')
      print('----------')
      print(resp)
    if (is_return_val):
      bdata = rdata
    return bdata
  # 
  # ask me function
  def ask_me(self, prompt, 
    model="text-davinci-003",
    suffix=None,
    max_tokens=128,    # length of output, max=2048
    temperature=1.0,  # randomness: 0 to 2.0
    top_p=1.0,        # accurate: 0 to 1.0
    n=1,              # number of output
    stream=False,     # partial progress return
    logprobs=None,    # log properbility of token
    echo=True,        # include the prompt in the response
    stop=None,        # stop process on this character
    presence_penalty=0,   # likelyhood of new topic: -2.0 to 2.0
    frequency_penalty=0,  # llikelyhood of repeat: -2.0 to 2.0
    best_of=1,        # best of choices from "n" above
    logit_bias=None,  # do not use this word
    user='None',      # user name for reporting back to OpenAI
    is_print_json=True,
    is_return_val=False
    ):
    response = openai.Completion.create(
      prompt=prompt,
      model=model,
      suffix=suffix,
      max_tokens=max_tokens,
      temperature=temperature,
      top_p=top_p,
      n=n,
      stream=stream,
      logprobs=logprobs,
      echo=echo,
      stop=stop,
      presence_penalty=presence_penalty,
      frequency_penalty=frequency_penalty,
      best_of=best_of,
      #logit_bias=logit_bias,
      user=user
    )
    self.print_answer(response,is_print_json=is_print_json,is_return_val=is_return_val)
    return
```
