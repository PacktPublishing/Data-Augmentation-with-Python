```python
import pandas as pd
import requests
import os

# updat to full path
df = pd.read_csv("/content/articles-pet.csv") #Upload the CSV file to session storage and update the path accordingly

df['local_link'] = ""

if not os.path.exists("articles_text_file"):
    os.mkdir("articles_text_file")

for index, row in df.iterrows():
    url = row['Article URL']
    response = requests.get(url)
    article_text = response.text
    text_filename = f"articles_text_file/article_{index}.text"
    save_directory = 'articles_text_file'
    with open(text_filename, 'w', encoding='utf-8') as file:
        file.write(article_text)
    df.at[index, 'local_link'] = os.path.join(text_filename)

df.to_csv("updated_data.csv", index=False)
```
